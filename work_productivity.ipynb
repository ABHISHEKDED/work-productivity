```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('remote_work_productivity.csv')
```

```python
data.head()
```

```python
data.info()
```

here just employment type is in object hence we will encode the data and mark them as 0,1 (0 for remote and 1 for in office)

```python
data['Employment_Type'].unique()
```

```python
data['Employment_Type']=data['Employment_Type'].map({'Remote':0,'In-Office':1})
```

```python
data.head()
```

```python
data.info()
```

```python
data['Productivity_Score'].unique()
```

```python
data['Productivity_Score'].max()
```

```python
data['Productivity_Score'].min()
```

```python
def map_values(x):
    if 0 <= x <= 40:
        return 0
    elif 41 <= x <= 70:
        return 1
    else:
        return 2

data['Productivity_Score'] = data['Productivity_Score'].apply(map_values)

data.head()
```

```python
data_productivity_0 = data[data['Productivity_Score'] == 0]
sns.countplot(x='Employment_Type', data=data_productivity_0)
plt.title('Employment Type Distribution for Productivity Score 0')
plt.xlabel('Employment Type')
plt.ylabel('Count')
plt.show()
```

```python
data_productivity_1 = data[data['Productivity_Score'] == 1]
sns.countplot(x='Employment_Type', data=data_productivity_1)
plt.title('Employment Type Distribution for Productivity Score 1')
plt.xlabel('Employment Type')
plt.ylabel('Count')
plt.show()
```

```python
data_productivity_2 = data[data['Productivity_Score'] == 2]
sns.countplot(x='Employment_Type', data=data_productivity_2)
plt.title('Employment Type Distribution for Productivity Score 2')
plt.xlabel('Employment Type')
plt.ylabel('Count')
plt.show()
```

```python
# prompt: create a countplot using seaborn between Employment_Type and Productivity_Score

sns.countplot(x='Employment_Type', hue='Productivity_Score', data=data,palette='Set2')
plt.title('Employment Type vs Productivity Score')
plt.xlabel('Employment Type')
plt.ylabel('Count')
plt.show()
```
